```python

classic_asset_data.info()
sns.pairplot(classic_asset_data)
```

```python
# Creating a list with the common quarters of both Databases

quarter_classic = classic_asset_data['QUARTER']
quarter_alternativ = alternative_asset_data['QUARTER']
mutual_quarter = set(quarter_classic) & set(quarter_alternativ)
quarter = sorted(list(mutual_quarter))

alternative_asset_data = alternative_asset_data[alternative_asset_data['QUARTER'] in quarter]
classic_asset_data  = classic_asset_data[classic_asset_data['QUARTER'] in quarter]

us_corporate_bond = classic_asset_data['USD Corporate Bond - USD Unhedged']
us_private_equity = alternative_asset_data['Private Equity USD Unhedged']
plt.figure(figsize = (12,6))
plt.plot(quarter, us_corporate_bond)
plt.plot(quarter, us_private_equity)
plt.xticks(rotation=45)
plt.xticks(ticks = quarter[::200])
plt.show()
```

```python
resultat = pd.merge(alternative_asset_data, classic_asset_data, on='QUARTER', how='inner')
resultat = resultat.dropna(subset = ['USD Corporate Bond - USD Unhedged'])
us_private_equity = resultat['Private Equity USD Unhedged']
us_corporate_bond = resultat['USD Corporate Bond - USD Unhedged']


plt.figure(figsize = (12,6))
plt.plot(resultat['QUARTER'], us_private_equity)
plt.plot(resultat['QUARTER'], us_corporate_bond)
plt.grid()
plt.xticks(rotation=45)
plt.xticks(ticks = quarter[::4])
plt.show()
```
